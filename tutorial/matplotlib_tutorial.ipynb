# Matplotlib Tutorial

```python
import matplotlib.pyplot as plt
```

```python
x_values = [1,2,3,4]
y_values = [5,6,7,8]

plt.plot(x_values, y_values)
plt.show()
```

```python
x_other_values = [1,2,3,4]
y_other_values = [5,1,7,4]

plt.scatter(x_other_values, y_other_values, color='navy')
plt.title('Scatter Sample Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
```
